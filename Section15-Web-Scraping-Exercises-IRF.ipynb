```python
## need to create a method to check if anything of value is happening

quotelist = [] ## create empty quote list

i = 1
while True:
    
    try:
        base_url = 'https://quotes.toscrape.com/page/{}/' ## define base url
        scrape_url = base_url.format(i) ## set scrape url to work off iterator
        
        ## create response for each url
        response = requests.get(scrape_url)
        
        ## beautifulsoup each response
        soup = bs4.BeautifulSoup(response.text, 'lxml') ## pass in response.text and parse type 'lxml'
        
        for item in soup.select('.quote > .text'):
            quotelist.append(item.getText())
            print(quotelist)
            print('\n')
        
        i += 1
        
        ## hacky way but working! while loop stops.
        if 'No quotes found!' in str(soup):
            break
    except: 
        break
```

```python
## lol FUCK- FOR AUTHORS NOT QUOTES

## need to create a method to check if anything of value is happening

authorslist = set() ## create empty quote list

i = 1
while True:
    
    try:
        base_url = 'https://quotes.toscrape.com/page/{}/' ## define base url
        scrape_url = base_url.format(i) ## set scrape url to work off iterator
        
        ## create response for each url
        response = requests.get(scrape_url)
        
        ## beautifulsoup each response
        soup = bs4.BeautifulSoup(response.text, 'lxml') ## pass in response.text and parse type 'lxml'
        
        for item in soup.select('.author'):
            authorslist.add(item.getText())
            print(authorslist)
            print('\n')
        
        i += 1
        
        ## hacky way but working! while loop stops.
        if 'No quotes found!' in str(soup):
            break
    except: 
        break
```
